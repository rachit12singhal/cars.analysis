### 1. What is name of car that has the highest horsepower?  

```python
#your code here
import pandas as pd


df = pd.read_csv('cars.csv')

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')


df_clean = df.dropna(subset=['horsepower'])

max_hp_row = df_clean.loc[df_clean['horsepower'].idxmax()]


car_name = max_hp_row['name']
horsepower = max_hp_row['horsepower']

print(f"The car with the highest horsepower is: {car_name} with {horsepower} hp")

```

### 2. How many cars have mpg  35?

```python
#your code here
import pandas as pd


df = pd.read_csv('cars.csv')


df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')


high_mpg_cars = df[df['mpg'] >= 35].dropna(subset=['mpg'])


count = len(high_mpg_cars)

print(f"Number of cars with mpg  35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
import pandas as pd


df = pd.read_csv('cars.csv')


df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')


high_hp_cars = df[df['horsepower'] > 100]


most_common_origin = high_hp_cars['origin'].mode()[0]

print(f"The most common origin for cars with horsepower > 100 is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
import pandas as pd

df = pd.read_csv('cars.csv')


df['acceleration'] = pd.to_numeric(df['acceleration'], errors='coerce')


japan_cars = df[df['origin'] == 'japan']


mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print(f"The mean acceleration of cars from Japan is: {mean_acceleration}")
```

### 5. Which year had the highest average mpg?

```python
import pandas as pd

df = pd.read_csv('cars.csv')
print(df.columns)

mean_mpg_by_year = df.groupby('model_year')['mpg'].mean()


best_year = mean_mpg_by_year.idxmax()
best_avg_mpg = mean_mpg_by_year.max()

print(f"The year with the highest average mpg is {best_year} with an average mpg of {best_avg_mpg:.2f}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
import pandas as pd


df = pd.read_csv('cars.csv')


df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

median_mpg = df['mpg'].median()


high_mpg_cars = df[df['mpg'] > median_mpg].copy()


high_mpg_cars['hp_to_weight'] = high_mpg_cars['horsepower'] / high_mpg_cars['weight']


max_ratio = high_mpg_cars['hp_to_weight'].max()


best_cars = high_mpg_cars[high_mpg_cars['hp_to_weight'] == max_ratio]


for _, row in best_cars.iterrows():
    print(f"Car: {row['name']}, Horsepower/Weight ratio: {row['hp_to_weight']:.6f}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('cars.csv')

# Convert columns to numeric (handle any bad data)
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')  # might be called 'year' in your dataset

# Drop rows with missing data needed for plot
df_clean = df.dropna(subset=['mpg', 'model_year', 'origin'])

# Group by model year and origin, then calculate mean mpg
avg_mpg = df_clean.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year', fontsize=14)
plt.ylabel('Average MPG', fontsize=14)
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('cars.csv')


df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')


df_clean = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])


hue_order = ['japan', 'europe', 'usa']


plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    data=df_clean,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),  # size range for mpg
    palette='Set1',
    hue_order=hue_order,
    alpha=0.7
)


plt.title('Horsepower vs Weight of Cars Colored by Origin and Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.legend(title='Origin', fontsize=12, title_fontsize=13, loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd


df = pd.read_csv('cars.csv')

df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')


appearances = df.groupby('name')['model_year'].nunique()


multi_year_cars = appearances[appearances >= 2].index

df_multi_year = df[df['name'].isin(multi_year_cars)]

stats = df_multi_year.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean'),
    appearances=('model_year', 'nunique')
).reset_index()


consistent_cars = stats[stats['mpg_std'] < 1.0]

min_std = consistent_cars['mpg_std'].min()
best_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

best_cars_sorted = best_cars.sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])


print(best_cars_sorted[['name', 'appearances', 'mpg_mean', 'mpg_std']])
```
